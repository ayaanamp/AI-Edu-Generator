```python
!pip install PyPDF2

import PyPDF2
import random
```

```python
def extract_text_from_pdf(file):
    text = ""
    reader = PyPDF2.PdfReader(file)

    for page in reader.pages:
        text += page.extract_text()

    return text
```

```python
def clean_text(text):
    return text.replace("\n", " ").strip()
```

```python
def generate_flashcards(text):
    sentences = text.split(".")
    cards = []

    for s in sentences[:10]:
        if len(s) > 20:
            cards.append(("Explain: " + s.strip(), s.strip()))

    return cards
```

```python
def get_key_points(text):
    sentences = text.split(".")
    return [s.strip() for s in sentences if len(s.split()) > 8][:10]
```

```python
def generate_mcq(text):
    sentences = text.split(".")
    mcqs = []

    for s in sentences[:5]:
        if len(s) > 20:
            correct = s.strip()
            options = [correct, "Option A", "Option B", "Option C"]
            random.shuffle(options)

            mcqs.append({
                "question": s.strip(),
                "options": options,
                "answer": correct
            })

    return mcqs
```

```python
from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

text = extract_text_from_pdf(file_name)
cleaned = clean_text(text)
```

```python
cards = generate_flashcards(cleaned)

for q, a in cards:
    print("Q:", q)
    print("A:", a)
    print()
```

```python
points = get_key_points(cleaned)

for p in points:
    print("-", p)
```

```python
mcqs = generate_mcq(cleaned)

for m in mcqs:
    print("Q:", m["question"])
    print("Options:", m["options"])
    print("Answer:", m["answer"])
    print()
```
